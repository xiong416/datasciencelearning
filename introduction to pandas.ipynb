```python
# column from series
seats_column = pd.Series([5,5,5,5,5])

# new column called Seats
car_sales["Seats"] = seats_column
```

```python
car_sales
```

```python
# fillna with 5
car_sales["Seats"].fillna(5, inplace=True)
```

```python
# column from Python List
feul_econamy = [7.5,6.7,9.2,5.8,4.7,8.4,6.4,7.2,9.2,6.8]
car_sales["Feul per KM"] = feul_econamy
car_sales
```

```python
# create customized column from existing columns
car_sales["Total feul used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Feul per KM"]
car_sales
```

```python
# create column from single value
car_sales["Number of wheels"] = 4
car_sales
```

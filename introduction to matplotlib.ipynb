### two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,5))

# plot for each
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))

# plot for each
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax[1,1].hist(np.random.randn(1000));
```

### which one should be used? (pyplot vs matplotlib OO method)
* when plot quickly, pyplot method
* when plot in a more advanced way, OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax = ax);
#ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots()

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#customize plots
ax.set(title="heart disease and cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol");

#legend
ax.legend(*scatter.legend_elements(), title="Target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head(5)
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# setting
ax0.set(title="Heart Disease and Cholestrol Levels",
       #xlabel="Age",
       ylabel="Cholestrol")

# add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# add meanline
ax0.axhline(over_50["chol"].mean(),
           linestyle = "--");


# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# setting
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# add legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add meanline
ax1.axhline(over_50["thalach"].mean(),
           linestyle = "--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = "16", fontweight = "bold");
```

```python
# set style
plt.style.use("seaborn-whitegrid")

# OO method from scratch
fig, ax = plt.subplots()

#plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

#customize plots
ax.set(title="heart disease and cholestrol levels",
      xlabel="Age",
      ylabel="Cholestrol");

#legend
ax.legend(*scatter.legend_elements(), title="Target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# customizing style
plt

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# setting
ax0.set(title="Heart Disease and Cholestrol Levels",
       #xlabel="Age",
       ylabel="Cholestrol")
ax0.set_xlim(50,80)

# add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# add meanline
ax0.axhline(over_50["chol"].mean(),
           linestyle = "--");


# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# setting
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])
# add legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add meanline
ax1.axhline(over_50["thalach"].mean(),
           linestyle = "--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = "16", fontweight = "bold");

# save the fig
fig.savefig("heart-disease-analysis-plot.png")
```
